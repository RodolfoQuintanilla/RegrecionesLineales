```python
url=r"https://raw.githubusercontent.com/bencmbit/datasets/refs/heads/master/wine.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(url)
df.head()
```

```python
plt.plot(df[["Alcohol", "Ash", "Mg", "Acl"]])
plt.title("Ejemplo Basico")
plt.xlabel("Eje X")
plt.ylabel("Eje y")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

df[["Alcohol", "Ash", "Mg", "Acl"]].boxplot()

plt.show()
```

```python
plt.plot(df[["Alcohol", "Ash", "Mg", "Acl"]],
            marker="x")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(df["Alcohol"], df["Ash"], df["Mg"],color="red", marker="x")

plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
plt.bar(df["Alcohol"], df["Ash"], df["Mg"], edgecolor="black")
plt.show()
```

```python
plt.hist(df["Alcohol"], bins=30, color="skyblue", edgecolor="black")
```

```python
plt.boxplot(df[["Alcohol", "Ash", "Mg", "Acl"]])
```

```python
labels = ["Alcohol", "Ash", "Flavanoids", "Mg"]

valores = [
    df["Alcohol"].mean(),
    df["Ash"].mean(),
    df["Flavanoids"].mean(),
    df["Mg"].mean()
]

plt.pie(
    valores,
    labels=labels,
    autopct="%.1f%%",
    startangle=90
)

plt.show()
```
