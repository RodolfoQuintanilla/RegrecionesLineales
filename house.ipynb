### EDA

```python
# Modificacion de los datos
df_new = df.copy()
```

```python
df_new.columns
```

```python
# Modifficar nombre de las columnas


df_new.rename(columns={
    'Index': 'indice',
    'Title': 'titulo',
    'Description': 'descripcion',
    'Amount(in rupees)': 'monto',
    'Price (in rupees)': 'precio',
    'location': 'ubicacion',
    'Carpet Area': 'area_util',
    'Status': 'estado',
    'Floor': 'piso',
    'Transaction': 'transaccion',
    'Furnishing': 'mobiliario',
    'facing': 'orientacion',
    'overlooking': 'vista',
    'Society': 'sociedad',
    'Bathroom': 'banos',
    'Balcony': 'balcones',
    'Car Parking': 'estacionamiento',
    'Ownership': 'propiedad',
    'Super Area': 'area_total',
    'Dimensions': 'dimensiones',
    'Plot Area': 'terreno'
}, inplace=True)
```

```python
df_new["propiedad"].value_counts()
```

```python
df_new.isnull().sum()
```

```python
df_new = df_new.drop(columns=["terreno", "dimensiones", "area_total","orientacion","transaccion","vista"])
```

```python
df_new.head(3)
```

```python
# valores nulos

df_new.isnull().sum()
```

```python
df_new["descripcion"] = df_new["descripcion"].fillna("Sin descripcion")
```

```python
print(f"porcentaje de valores nulos en precio: {df_new["precio"].isnull().sum() /len(df_new)*100:.2f}%")
```

```python
# imputacion de precio
df_new["precio"] = df_new["precio"]
```

```python
df_new["precio"].fillna(df_new["precio"].mean(), inplace=True)
```

```python
df_new["area_util"] = df_new["area_util"].str.replace("sqft", "",regex= False) 
df_new["area_util"] = pd.to_numeric(df_new["area_util"], errors="coerce")
```

```python
df_new["area_m2"] = df_new["area_util"] * 0.092903

df_new["area_m2"].head(3)
```

```python
df_new["area_m2"] = df_new["area_m2"].fillna(
    df_new.groupby("titulo")["area_m2"]
    .transform("mean")
)
```

```python
df_new["area_m2"].fillna(df_new["area_m2"].mean(), inplace=True)
```

```python
df_new.drop(columns=["area_util","piso","sociedad","estacionamiento", "propiedad" ], inplace=True)
```

```python
df_new["estado"].fillna("Desconocido", inplace=True)
```

```python
df_new["mobiliario"].fillna("Desconocido", inplace=True)
```

```python
df_new["banos"]= pd.to_numeric(df_new["banos"], errors="coerce")
```

```python
df_new["banos"].fillna(df_new["banos"].median(), inplace=True)
```

```python
df_new["balcones"]= pd.to_numeric(df_new["balcones"], errors="coerce")
```

```python
df_new["balcones"].fillna(df_new["balcones"].median(), inplace=True)
```
